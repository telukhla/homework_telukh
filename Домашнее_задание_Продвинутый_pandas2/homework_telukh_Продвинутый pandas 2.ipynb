# Задание 1

```python
ratings = pd.read_csv('/Users/larisatelukh/Documents/Product_Analytics/Python/доп_материалы_к_пандас_1_и_2/ml-latest-small/ratings.csv')

ratings.head()
```

```python
# конвертации из UNIX time

ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')
```

```python
# группируем по пользователям, считаем общее число оценок и вычисляем max и min значение столбца timestamp

results = ratings.groupby('userId').agg({'rating': 'count', 'timestamp': ['min', 'max']})

results.head()
```

```python
# убираем multidimensional index 

results.columns = [''.join(tuple(map(str, t))) for t in results.columns.values]

results
```

```python
# вычисляем среднее время жизни пользователя

results['average'] = results['timestampmax'] - results['timestampmin']

results
```

```python
# Фильтруем пользователей, которые выставили более 100 оценок и вычисляем среднее время жизни

results.loc[results.ratingcount >= 100].average.mean()
```

Дана статистика услуг перевозок клиентов компании по типам:

rzd - железнодорожные перевозки

auto - автомобильные перевозки

air - воздушные перевозки

client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)

rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)

auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)

air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)

client_base
```

### таблицу с тремя типами выручки для каждого client_id без указания адреса клиента



```python
# создаем результирующую таблицу с client_id

results = pd.DataFrame(client_base.loc[:,'client_id'])

results
```

```python
# Добавляем три типа выручки для каждого client_id

results.merge(rzd, how='left', on='client_id').merge(auto, how='left', on='client_id').merge(air, how='left', on='client_id')
```

### аналогичную таблицу по типам выручки с указанием адреса клиента



```python
results2 = client_base.merge(rzd, how='left', on='client_id').merge(auto, how='left', on='client_id').merge(air, how='left', on='client_id')

results2
```
