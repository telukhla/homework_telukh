```python
def movie_classification(rating):
    """функция, которая классифицирует фильмы из материалов занятия по следующим правилам:
        - оценка 2 и меньше - низкий рейтинг
        - оценка 4 и меньше - средний рейтинг
        - оценка 4.5 и 5 - высокий рейтинг"""
    classification = ''
    
    if rating <= 2:
        classification = "низкий рейтинг"
    elif  2 < rating <=  4:
        classification = "средний рейтинг"
    elif 4 < rating <=  5:
        classification = "высокий рейтинг"
        
    return classification
```

```python
movies['class'] = movies['rating'].apply(movie_classification)

movies.head(30)
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
import re
```

```python
# не проверяются окончания слов 

def define_region(value):
    
    temp_list = re.findall(r'\w+', value)
   
    
    for i in temp_list: 
        for region, towns in geo_data.items():
            if i in towns:
                return region
            else: 
                return "undefined"
```

```python
data['region'] = data['keyword'].apply(define_region)

data.head()
```

```python
data = pd.read_csv('/Users/larisatelukh/Documents/Product_Analytics/Python/ml-latest-small-from-lecture-Pandas/ratings.csv')

data.head()
```

```python
movies = pd.read_csv('/Users/larisatelukh/Documents/Product_Analytics/Python/ml-latest-small-from-lecture-Pandas/movies.csv')

movies.head()
```

```python
# В переменную years записан список из всех годов с 1950 по 2020.

years = list(range(1950, 2021))

print(years)
```

```python
def production_year(name):
    """
    1) проверяем, что год в списке years
    2) если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    3) если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
    """
    
    list_year = re.findall(r'\d{4}', name)
    
    if (list_year != []):
        for i in list_year:
            if int(i) in years:
                return i
            else:
                return '1900'
    else:
        return '1900'
```

```python
# проверка на одном названии фильма

production_year('Toy Story 1936')
```

```python
# Записываем год выпуска фильма по алгоритму функции production_year в новый столбец ‘year’

movies['year'] = movies['title'].apply(production_year)

movies
```

```python
# выводим средний рейтинг каждого фильма

movies['rating'] = data.groupby('movieId').mean().reset_index()[['rating']]

movies
```

```python
# Считаем средний рейтинг всех фильмов для каждого значения столбца ‘year’ и 
# отсортируем результат по убыванию рейтинга

movies.groupby('year').mean().reset_index()[['year','rating']].sort_values('rating', ascending=False).head(30)
```
