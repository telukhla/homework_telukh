### Проанализируем разные столбцы

```python
# surgery - выбросов нет, есть 1 незаполненное значение
```

```python
# Hospital_number - что-то не так со значениями, приведем все к numeric

df['Hospital_Number'] = df['Hospital_Number'].apply(lambda x: int(x))
```

```python
df.describe()
```

```python
# найдем выбросы через межквартльный размах (на примере возраста)

q1 = df.Age.quantile(0.25)
q3 = df.Age.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df.Age.between(lower_bound, upper_bound, inclusive=True)].sort_values('Age')
remove_outliers
```

```python
# что это за выброс?
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
```

```python
# Age - есть лошадки с возрастом 9, хотя из описания данное поле может принимать только 2 значения, 
# скорее всего это ошибка заполнения, можно проставить 2 ?
```

```python
df['Age'] = df['Age'].apply(lambda x:2 if x>1 else 1)
```

```python
df.describe()
```

```python
# А теперь найдем выбросы через межквартльный размах (на примере пульса)


q1 = df.Age.quantile(0.25)
q3 = df.Age.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df.pulse.between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```
